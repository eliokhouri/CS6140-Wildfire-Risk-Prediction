```python
import csv
import json
from google.colab import drive
drive.mount('/content/drive')
```

```python
INFILE_NAME = "/content/drive/My Drive/ML6140  - Project/Data/basic_data.csv"
OUTFILE_NAME = "/content/drive/My Drive/ML6140  - Project/forMappingPoints/pointsToMap.json"
```

```python
positiveExamples = []
negativeExamples = []
with open(INFILE_NAME, "r", newline="") as infile:
  reader = csv.DictReader(infile)
  for row in reader:
    try:
      if int(row["class_label"]):
        positiveExamples.append([float(row["incident_longitude"]), float(row["incident_latitude"])])  # lon, lat for GeoJSON
      else:
        negativeExamples.append([float(row["incident_longitude"]), float(row["incident_latitude"])])
    except ValueError:
      pass
print("len(positiveExamples): ", len(positiveExamples))
print("len(negativeExamples): ", len(negativeExamples))  # That's a problem!!
```

```python
ghcnStations = []
with open("/content/drive/My Drive/ML6140  - Project/Raw Data/NOAA_GHCN-Daily/melio-ca-stations.csv", "r", newline="") as infile:
  reader = csv.reader(infile)
  for row in reader:
    ghcnStations.append([float(row[2]), float(row[1])])
print("len(ghcnStations): ", len(ghcnStations))
```

```python
lcdStations = []
with open("/content/drive/My Drive/ML6140  - Project/Raw Data/LCD_STATION_DATA/LCD_MY_STATIONS_ALL_INFO.csv", "r", newline="") as infile:
  reader = csv.DictReader(infile)
  for row in reader:
    lcdStations.append([float(row["STATION_LONGITUDE"]), float(row["STATION_LATITUDE"])])
print("len(lcdStations): ", len(lcdStations))
```

```python
outObj = {
    "positiveExamples": positiveExamples,
    "negativeExamples": negativeExamples,
    "ghcnStations": ghcnStations,
    "lcdStations": lcdStations
}
with open(OUTFILE_NAME, "w") as outfile:
  outfile.write(json.dumps(outObj))
```
